```python
# 遍历文件夹中的所有文件

images = {}

image_folder = "results\\facades_pix2pix\\test_latest\images"

for filename in os.listdir(image_folder):
    if filename.endswith("_fake_B.png") or filename.endswith("_real_B.png"):
        # 获取基本文件名（不包含_fake_B或_real_B）
        base_name = filename.split('_')[0]
        if base_name not in images:
            images[base_name] = []
        images[base_name].append(os.path.join(image_folder, filename))
```

```python
images
```

```python
# 遍历字典，计算每对图片的PSNR
psnr_list = []

for base_name, file_paths in images.items():
    if len(file_paths) == 2:
        # 加载图片
        img1 = np.array(Image.open(file_paths[0]))
        img2 = np.array(Image.open(file_paths[1]))
        # 计算 PSNR
        psnr = compare_psnr(img1, img2)
        psnr_list.append(psnr)
        print(f"PSNR between {os.path.basename(file_paths[0])} and {os.path.basename(file_paths[1])}: {psnr}")
```

```python
psnr_list
```

```python
# 计算 PSNR 值的平均值
average_psnr = sum(psnr_list) / len(psnr_list)

average_psnr
```
